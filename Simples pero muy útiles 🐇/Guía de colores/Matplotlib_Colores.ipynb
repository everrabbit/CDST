# Colores - Matplotlib

```python
import matplotlib.colors as mcolors
sorted_color_names = sorted(mcolors.CSS4_COLORS, key=lambda c: tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(c))))
def crear_tabla_completa_colores(filas=37, mostrar_directamente=True):
    """
    Crea una tabla completa con todos los colores organizados en el n√∫mero de filas especificado
    
    Par√°metros:
    -----------
    filas : int
        N√∫mero de filas para organizar la tabla (default: 37)
    mostrar_directamente : bool
        Si True, muestra el resultado renderizado. Si False, devuelve el string Markdown
    """
    global sorted_color_names
    
    if not sorted_color_names:
        print("‚ö†Ô∏è  Primero ejecuta la celda que crea sorted_color_names")
        return "‚ö†Ô∏è Lista de colores no disponible"
    
    import matplotlib.colors as mcolors
    import math
    
    total_colores = len(sorted_color_names)
    columnas = math.ceil(total_colores / filas)
    
    print(f"üìä Organizando {total_colores} colores en {filas} filas y {columnas} columnas")
    
    # Crear encabezados de la tabla (sin columna de fila)
    headers = [f"Color {i+1}" for i in range(columnas)]
    
    # Crear la tabla Markdown con estilo mejorado (compacto)
    markdown_output = f"## Tabla Completa de Colores ({filas} filas √ó {columnas} columnas)\n\n"
    
    # CSS compacto
    markdown_output += """<style>
.color-table {width: 100%; border-collapse: collapse; font-family: Arial, sans-serif;}
.color-table th, .color-table td {text-align: center; vertical-align: middle; padding: 8px 6px; border: 1px solid #ddd; min-width: 74px;}
.color-table th {background-color: #f5f5f5; font-weight: bold; font-size: 13px;}
.color-table td {font-size: 12px;}
.color-block {display: inline-block; width: 22px; height: 22px; border-radius: 3px; margin-bottom: 4px; border: 1px solid #ccc;}
.color-name {font-weight: bold; font-size: 11px; display: block; margin-top: 2px;}
</style>
<table class="color-table"><thead><tr>""" + "".join([f"<th>{header}</th>" for header in headers]) + """</tr></thead><tbody>"""
    
    # Llenar la tabla (sin columna de fila)
    for fila in range(filas):
        markdown_output += "<tr>"
        
        for col in range(columnas):
            indice = fila + (col * filas)
            
            if indice < total_colores:
                color_name = sorted_color_names[indice]
                
                # Obtener c√≥digo hex
                if color_name in mcolors.CSS4_COLORS:
                    hex_color = mcolors.CSS4_COLORS[color_name]
                else:
                    hex_color = '#808080'  # Gris por defecto
                
                # Crear celda compacta
                celda_contenido = f"""<div style="text-align: center;"><div class="color-block" style="background-color: {hex_color}; display: inline-block;"></div><div class="color-name" style="color: {hex_color};">{color_name}</div><div style="font-size: 10px; color: #666; margin-top: 2px;">{hex_color}</div></div>"""
            else:
                celda_contenido = "<div style='text-align: center; color: #ccc;'>-</div>"
            
            markdown_output += f"<td>{celda_contenido}</td>"
        
        markdown_output += "</tr>"
    
    markdown_output += "</tbody></table>"
    
    if mostrar_directamente:
        from IPython.display import Markdown, display
        display(Markdown(markdown_output))
    else:
        return markdown_output
```

```python
crear_tabla_completa_colores(filas=37)
```
