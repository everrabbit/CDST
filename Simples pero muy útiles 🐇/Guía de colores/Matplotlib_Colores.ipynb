```python
import matplotlib.colors as mcolors
sorted_color_names = sorted(mcolors.CSS4_COLORS, key=lambda c: tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(c))))
def crear_tabla_completa_colores(filas=37, mostrar_directamente=True):
    """
    Crea una tabla completa con todos los colores organizados en el n√∫mero de filas especificado
    
    Par√°metros:
    -----------
    filas : int
        N√∫mero de filas para organizar la tabla (default: 37)
    mostrar_directamente : bool
        Si True, muestra el resultado renderizado. Si False, devuelve el string Markdown
    """
    global sorted_color_names
    
    if not sorted_color_names:
        print("‚ö†Ô∏è  Primero ejecuta la celda que crea sorted_color_names")
        return "‚ö†Ô∏è Lista de colores no disponible"
    
    import matplotlib.colors as mcolors
    import math
    
    total_colores = len(sorted_color_names)
    columnas = math.ceil(total_colores / filas)
    
    print(f"üìä Organizando {total_colores} colores en {filas} filas y {columnas} columnas")
    
    # Crear encabezados de la tabla (sin columna de fila)
    headers = [f"Color {i+1}" for i in range(columnas)]
    
    # Crear la tabla Markdown con estilo mejorado (compacto)
    markdown_output = f"## Tabla Completa de Colores ({filas} filas √ó {columnas} columnas)\n\n"
    
    # CSS compacto con texto m√°s grande
    markdown_output += """<style>
.color-table {width: 100%; border-collapse: collapse; font-family: Arial, sans-serif;}
.color-table th, .color-table td {text-align: center; vertical-align: middle; padding: 12px 8px; border: 1px solid #ddd; min-width: 90px;}
.color-table th {background-color: #f5f5f5; font-weight: bold; font-size: 16px;}
.color-table td {font-size: 14px;}
.color-block {display: inline-block; width: 28px; height: 28px; border-radius: 4px; margin-bottom: 6px; border: 2px solid #333;}
.color-name {font-weight: bold; font-size: 14px; display: block; margin-top: 4px;}
.color-hex {font-size: 12px; font-family: monospace; margin-top: 3px;}
</style>
<table class="color-table"><thead><tr>""" + "".join([f"<th>{header}</th>" for header in headers]) + """</tr></thead><tbody>"""
    
    # Llenar la tabla (sin columna de fila)
    for fila in range(filas):
        markdown_output += "<tr>"
        
        for col in range(columnas):
            indice = fila + (col * filas)
            
            if indice < total_colores:
                color_name = sorted_color_names[indice]
                
                # Obtener c√≥digo hex
                if color_name in mcolors.CSS4_COLORS:
                    hex_color = mcolors.CSS4_COLORS[color_name]
                else:
                    hex_color = '#808080'  # Gris por defecto
                
                # Crear celda sin emojis
                celda_contenido = f"""<div style="text-align: center;">
                <div class="color-block" style="background-color: {hex_color}; display: inline-block;"></div>
                <div class="color-name" style="color: {hex_color}; font-size: 15px;"><strong>{color_name}</strong></div>
                <div class="color-hex" style="color: #666; font-size: 13px;">{hex_color}</div>
                </div>"""
            else:
                celda_contenido = "<div style='text-align: center; color: #ccc;'>-</div>"
            
            markdown_output += f"<td>{celda_contenido}</td>"
        
        markdown_output += "</tr>"
    
    markdown_output += "</tbody></table>"
    
    if mostrar_directamente:
        from IPython.display import Markdown, display
        display(Markdown(markdown_output))
    else:
        return markdown_output
```

```python
import math

import matplotlib.pyplot as plt

import matplotlib.colors as mcolors
from matplotlib.patches import Rectangle


def plot_colortable(colors, *, ncols=5, sort_colors=True):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        names = sorted(
            colors, key=lambda c: tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(c))))
    else:
        names = list(colors)

    n = len(names)
    nrows = math.ceil(n / ncols)

    width = cell_width * ncols + 2 * margin
    height = cell_height * nrows + 2 * margin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-margin)/height)
    ax.set_xlim(0, cell_width * ncols)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.add_patch(
            Rectangle(xy=(swatch_start_x, y-9), width=swatch_width,
                      height=18, facecolor=colors[name], edgecolor='0.7')
        )

    return fig
plot_colortable(mcolors.CSS4_COLORS)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import random

def graficar_barras_por_lotes(lote=30):
    """
    Crea gr√°ficos de barras usando la lista ordenada de colores HSV
    """
    global sorted_color_names
    
    if not sorted_color_names:
        print("‚ö†Ô∏è  Primero ejecuta la celda que crea sorted_color_names")
        return
    
    nombres = sorted_color_names  # Usar la lista ordenada por HSV
    total = len(nombres)
    lotes = [nombres[i:i + lote] for i in range(0, total, lote)]

    for i, grupo in enumerate(lotes, 1):
        valores = [random.randint(1, 10) for _ in grupo]
        # Obtener colores usando la misma l√≥gica que crear_tabla_completa_colores
        colores_lote = [mcolors.CSS4_COLORS[nombre] for nombre in grupo]

        fig, ax = plt.subplots(figsize=(max(12, len(grupo) * 0.6), 6))
        barras = ax.bar(grupo, valores, color=colores_lote, edgecolor='#e0e0e0')

        for bar, name in zip(barras, grupo):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width() / 2, height + 0.3,
                    name, ha='center', va='bottom', rotation=90, fontsize=18)

        ax.set_title(f'Colores del {((i - 1) * lote) + 1} al {min(i * lote, total)} (Ordenados por HSV)',
                     fontsize=14, weight='bold')
        ax.set_xticks([])
        ax.set_ylim(0, 20)
        ax.set_facecolor('#f9f9f9')
        fig.patch.set_facecolor('#f9f9f9')
        plt.tight_layout()
        plt.show()

# Ejecutar con los colores ordenados por HSV
graficar_barras_por_lotes(lote=30)
```
