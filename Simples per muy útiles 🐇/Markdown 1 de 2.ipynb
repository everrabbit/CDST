# 📝 Niveles de texto en Markdown

¡Bienvenido a tu primer hechizo para que tus notas y proyectos se vean bien organizados!  
En Markdown, los **títulos y subtítulos** se crean usando el símbolo `#`.  
Mientras más `#` pongas, **más pequeño y específico** es el nivel del texto.

---

## 🏆 Niveles de encabezado

| Nivel | Sintaxis       | Resultado esperado                 | Uso sugerido                        |
|-------|----------------|------------------------------------|-------------------------------------|
| 1     | `# Título`     | 🟦 Grande, importante              | Nombre del proyecto / tema general |
| 2     | `## Subtítulo` | 🟩 Mediano                         | Secciones principales               |
| 3     | `### Tercer nivel` | 🟨 Más pequeño               | Subtemas o apartados               |
| 4     | `#### Cuarto`  | 🟧 Aún más pequeño                 | Detalles dentro de un subtema       |
| 5+    | `##### Quinto` | 🟥 Apenas se nota, ¡úsalo con cuidado! | Casi nunca necesario               |

---

## ✨ Ejemplo mágico

```markdown
# 🧠 Proyecto de Ciencia de Datos

## 📊 Análisis de ventas

### 🔍 Datos utilizados
- Excel con historial de ventas 2022-2024
- SQL para consultas de clientes
- Python para gráficas

### 📈 Resultados preliminares

#### Total de ventas por mes
(Gráfico generado con Matplotlib)

#### Clientes con mayor frecuencia
(Tabla generada con Pandas)

## 🐇 Conclusiones
- Las ventas crecen más en primavera
- El 20% de los clientes hacen el 80% del volumen
```
👁️ Se ve así

## 🔢 Listas en Markdown — Porque ver claro es pensar claro

### 🎯 ¿Por qué usar listas?

- Organizan ideas
- Sirven como checklist o índice de pasos
- Le dan respiro visual a tu texto
- Ayudan a resumir lo importante
- Se ven bien en GitHub, Notion, Jupyter, etc.

---

### ✅ Viñetas (listas con guiones o asteriscos)

```markdown
- Este es un punto simple
- Puedes tener subtareas:
  - Subnivel uno
    - Subnivel dos
* También puedes usar asteriscos
* Son iguales visualmente
- Puedes mezclar guiones y asteriscos
```
### 👁️ Se ve así
- Este es un punto simple
- Puedes tener subtareas:
  - Subnivel uno
    - Subnivel dos
* También puedes usar asteriscos
* Son iguales visualmente
- Puedes mezclar guiones y asteriscos
### 🔢 Listas numeradas (útiles para pasos o instrucciones)

```markdown
1. Abre VSCode
2. Escribe tu código
3. Corre el script
4. ¡Presume tus resultados!
```
### 👁️ Se ve así
1. Abre VSCode
2. Escribe tu código
3. Corre el script
4. ¡Presume tus resultados!

## 📌 Tips rápidos de estilo

- ✅ **Usa viñetas** si el orden _no importa_  
  (como para listar conceptos, herramientas, ideas, etc.)

- 🔢 **Usa números** si el orden _sí importa_  
  (como una receta, un tutorial paso a paso o el flujo de un análisis)

- ➕ **Puedes mezclar listas y subtítulos** para crear **índices o menús navegables**  
  (ideal para secciones de cursos o rutas de aprendizaje)

- 🐇 **Combina listas con emojis** para dar más intención visual  
  (el conejo aprueba esto 🐇✨)

# ¿Cómo ponemos color a nuestro Markdown?
## 🎨 Colores en Markdown
En Markdown puro no puedes cambiar colores directamente, pero puedes usar HTML para darle un toque especial a tus textos. Aquí te muestro cómo:

### 📝 **Qué puedes cambiar:**
- 🎨 **Color**: Cambia `#8B4513` por cualquier código de color hexadecimal
- ✏️ **Texto**: Reemplaza el contenido entre `>` y `<` (ejemplo: `"Este texto es marrón suave"`)

```markdown
<span style="color: #8B4513;">Este texto es marrón suave</span>
<span style="color: #4682B4;">Este texto es azul acero</span>
<span style="color: #6B8E23;">Este texto es verde oliva</span>
```
### 👁️ Se ve así
<span style="color: #8B4513;">Este texto es marrón suave</span>
<span style="color: #4682B4;">Este texto es azul acero</span>
<span style="color: #6B8E23;">Este texto es verde oliva</span>

---

## 🖌️ Colores en Markdown con HTML y CSS

### 📝 **Qué puedes cambiar:**
- 🎨 **Color**: Modifica los códigos de color (`#B8860B`, `#20B2AA`, `#9370DB`)
- ✏️ **Texto**: Cambia el contenido entre `>` y `<` (ejemplo: `"Texto en negrita"`)
- 🎯 **Estilo**: Usa `font-weight: bold` (negrita), `font-style: italic` (cursiva), `text-decoration: underline` (subrayado)

```markdown
<span style="color: #B8860B; font-weight: bold;">Texto en negrita</span>
<span style="color: #20B2AA; font-style: italic;">Texto en cursiva</span>
<span style="color: #9370DB; text-decoration: underline;">Texto subrayado</span>
```
### 👁️ Se ve así
<span style="color: #B8860B; font-weight: bold;">Texto en negrita</span>
<span style="color: #20B2AA; font-style: italic;">Texto en cursiva</span>
<span style="color: #9370DB; text-decoration: underline;">Texto subrayado</span>

---

## 🖌️ Colores en Markdown con HTML, CSS y emojis

### 📝 **Qué puedes cambiar:**
- 🎨 **Color**: Ajusta los códigos (`#A0522D`, `#228B22`, `#4682B4`)
- 🦊 **Emoji**: Cambia `🦊`, `🐰`, `🐳` por los emojis que prefieras
- ✏️ **Texto**: Modifica el contenido entre `>` y `<` después del emoji (ejemplo: `"Texto ejemplo"`)

```markdown
<span style="color: #A0522D;">🦊 Texto ejemplo</span>
<span style="color: #228B22;">🐰 Texto ejemplo</span>
<span style="color: #4682B4;">🐳 Texto ejemplo</span>
```
### 👁️ Se ve así
<span style="color: #A0522D;">🦊 Texto ejemplo</span>
<span style="color: #228B22;">🐰 Texto ejemplo</span>
<span style="color: #4682B4;">🐳 Texto ejemplo</span>

## 🖌️ Colores en Markdown con HTML, CSS, emojis, negritas y subrayados

### 📝 **Qué puedes cambiar:**
- 🎨 **Color**: Modifica `#DAA520`, `#5F9EA0`, `#8A2BE2`
