## 🖌️ Colores en Markdown con HTML, CSS, emojis, negritas, subrayados y tamaños

### 📝 **Qué puedes cambiar:**
- 🎨 **Color**: Ajusta `#D2691E`, `#2E8B57`, `#483D8B`
- 📏 **Tamaño**: Modifica `font-size: 20px`, `18px`, `16px` (puedes usar cualquier número)
- 🦊 **Emoji**: Cambia por los que prefieras
- ✏️ **Texto**: Personaliza `"Texto grande"`, `"Texto mediano"`, etc.

```markdown
<span style="color: #D2691E; font-weight: bold; text-decoration: underline; font-size: 20px;">🦊 Texto grande</span>
<span style="color: #2E8B57; font-style: italic; text-decoration: underline; font-size: 18px;">🐰 Texto mediano</span>
<span style="color: #483D8B; text-decoration: underline; font-weight: bold; font-size: 16px;">🐳 Texto pequeño</span>
```
### 👁️ Se ve así
<span style="color: #D2691E; font-weight: bold; text-decoration: underline; font-size: 20px;">🦊 Texto grande</span>
<span style="color: #2E8B57; font-style: italic; text-decoration: underline; font-size: 18px;">🐰 Texto mediano</span>
<span style="color: #483D8B; text-decoration: underline; font-weight: bold; font-size: 16px;">🐳 Texto pequeño</span>

## 🖌️ Colores en Markdown con HTML, CSS, emojis, negritas, subrayados, tamaños y alineación

### 📝 **Qué puedes cambiar:**
- 🎨 **Color**: Personaliza los códigos de color
- 📏 **Tamaño**: Ajusta los valores de `font-size`
- 📍 **Alineación**: Usa `text-align: center` (centro), `left` (izquierda), `right` (derecha)
- 🦊 **Emoji**: Sustituye por tus favoritos
- ✏️ **Texto**: Cambia los mensajes

# Colores de fondo en Markdown
## 🖌️ Colores de fondo en Markdown
En Markdown puro no puedes cambiar colores de fondo directamente, pero puedes usar HTML para darle un toque especial a tus textos. Aquí te muestro cómo:
```markdown
<div style="background-color: #F0E68C; padding: 10px; border-radius: 5px;">
  Este texto tiene un fondo amarillo claro.
</div>
<div style="background-color: #ADD8E6; padding: 10px; border-radius: 5px;">
  Este texto tiene un fondo azul claro.
</div>
<div style="background-color: #90EE90; padding: 10px; border-radius: 5px;">
  Este texto tiene un fondo verde claro.
</div>
```
### 👁️ Se ve así
<div style="background-color: #F0E68C; padding: 10px; border-radius: 5px;">
  Este texto tiene un fondo amarillo claro.
</div>
<div style="background-color: #ADD8E6; padding: 10px; border-radius: 5px;">
  Este texto tiene un fondo azul claro.
</div>
<div style="background-color: #90EE90; padding: 10px; border-radius: 5px;">
  Este texto tiene un fondo verde claro.
</div>
